```python
from sklearn.datasets import load_iris

# Returns a Bunch object, which is very similar to a dictionary. It contains keys and values
iris_dataset = load_iris()

# Print the keys of the dataset
print("Keys of iris_dataset:")
print(iris_dataset.keys())
```

```python
# Print description of the dataset
print("\nDescription of iris_dataset:")
print(iris_dataset['DESCR'][:193] + "\n...")  # Print

# print all of the description  
print(iris_dataset['DESCR'])
```

```python
# print the target names
print("Target names:")
print(iris_dataset['target_names'])

# print the feature names
print("\nFeature names:")
print(iris_dataset['feature_names'])

# print shape of data
print("\nShape of data:")
print(iris_dataset['data'].shape)

# print first 10 rows of data
print("\nFirst 10 rows of data:")
print(iris_dataset['data'][:10])

# print target values
print("\nFTarget values:")
print(iris_dataset['target'])
```
