```python
from sklearn.linear_model import ElasticNet, Ridge, Lasso

import numpy as np
from sklearn.model_selection import train_test_split
import mglearn

X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)


```

```python
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5).fit(X_train, y_train)
ridge = Ridge(alpha=0.5).fit(X_train, y_train)
lasso = Lasso(alpha=0.1).fit(X_train, y_train)

print("Elastic Net Training set score: {:.2f}".format(elastic_net.score(X_train, y_train)))
print("Ridge training set score: {:.2f}".format(ridge.score(X_train, y_train)))
print("Lasso training set score: {:.2f}".format(lasso.score(X_train, y_train)))
print("\n")
print("Elastic NetTest set score: {:.2f}".format(elastic_net.score(X_test, y_test)))
print("Ridge test set score: {:.2f}".format(ridge.score(X_test, y_test)))
print("Lasso test set score: {:.2f}".format(lasso.score(X_test, y_test)))
print("\n")
print("Number of features used in Elastic Net:", np.sum(elastic_net.coef_ != 0))
print("Number of features used in Ridge:", np.sum(ridge.coef_ != 0))
print("Number of features used in Lasso:", np.sum(lasso.coef_ != 0))
```
