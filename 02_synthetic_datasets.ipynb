```python
import mglearn
import matplotlib.pyplot as plt

# Create a synthetic dataset
# example of a synthetic two-class classification dataset is the forge dataset, which has two features.
X, y = mglearn.datasets.make_forge()
print("X.shape:", X.shape)
print("y.shape:", y.shape)

print("X : ", X)
print("X[:, 0] : ", X[:, 0])
print("X[:, 1] : ", X[:, 1])
print("y : ", y)

# plot dataset
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape:", X.shape)
```

```python
# To illustrate regression algorithms, we will use the synthetic wave dataset. 
# The wave dataset has a single input feature and a continuous target variable (or response) 
# that we want to model

X, y = mglearn.datasets.make_wave()

print("X.shape:", X.shape)
print("y.shape:", y.shape)

print("X : ", X)
print("y : ", y)

plt.plot(X, y, "o")
plt.xlabel("Feature")
plt.ylabel("Target")
```
